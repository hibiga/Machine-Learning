```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
def traitement(file) :
    for file in os.listdir("Machine Learning") :
        if file == "batches.meta" :
            continue
        with open(file, 'rb') as fo :
            dict = pickle.load(fo, encoding='bytes')

        data = [dict[b'data']].reshape((len(dict['data']), 3, 32, 32)).transpose(0, 2, 3, 1)
        label = [dict[b'labels']]

        return data, label

```
